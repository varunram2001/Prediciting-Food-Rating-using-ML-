# TIME TO PRE PROCESS CATEGORICAL COLUMNS


```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()

#Removing Recipe_Review alone as We will try to use tf vectorizer for that
categorical_columns.remove('Recipe_Review')

for i in categorical_columns:
    data_tt[i + '_L_Encoded'] = label_encoder.fit_transform(data_tt[i])
    data_val[i + '_L_Encoded'] = label_encoder.fit_transform(data_val[i])
    
    #We will drop the Encoded Columns
    data_tt.drop(i, axis=1, inplace=True)
    data_val.drop(i, axis=1, inplace=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()

#TFIDF Vectorizer for Recipe Review as it is a Long String
tfidf.fit(data_tt['Recipe_Review'])


recipe_review_tfidf_train = tfidf.transform(data_tt['Recipe_Review'])
recipe_review_tfidf_val = tfidf.transform(data_val['Recipe_Review'])

tfidf_train_df = pd.DataFrame(recipe_review_tfidf_train.toarray(), columns=tfidf.get_feature_names_out())
tfidf_val_df = pd.DataFrame(recipe_review_tfidf_val.toarray(), columns=tfidf.get_feature_names_out())

# Concatenate TF-IDF features with original datasets
data_tt = pd.concat([data_tt.reset_index(drop=True), tfidf_train_df], axis=1)
data_val = pd.concat([data_val.reset_index(drop=True), tfidf_val_df], axis=1)
```

```python
#TFIDF Vectorizer has been applied and new columns generated, main String column can be dropped
data_tt.drop(columns=['Recipe_Review'], inplace=True)
data_val.drop(columns=['Recipe_Review'], inplace=True)
```

```python
print("Shape of the Validation Set at this stage:", data_val.shape)
print("Shape of the Actual Set at this stage:", data_tt.shape)
```

```python
#Separating the Data given into the features and target Variable
Features_data = data_tt.drop(columns=['Rating'])
Target_data = data_tt['Rating']

#print("Target Variable Separated")
```

```python
#Checking the shape once again for confirmation
print("Shape of Features Set:", Features_data.shape)
print("Shape of the target varible data: ",Target_data.shape)
```

```python
#We will split the data given into Train and Test using the train_test_split library
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(Features_data,Target_data,test_size=0.2,random_state=42)
#print("Splitting Done")
```

```python
#These are the Data Shapes after Splitting
print("X_train Shape: ",X_train.shape)
print("y_train Shape: ", y_train.shape)
print("X_test Shape: ", X_test.shape)
print("y_test Shape: ", y_test.shape)
```
