```python
from sklearn.preprocessing import LabelEncoder

```

```python
dt = data_train.dtypes
text_col = dt[dt == 'object'].index.tolist()
for col in text_col:
    if len(data_train[col].unique()) <= len(data_train) / 2:
        data_train[col] = pd.factorize(data_train[col])[0]
    else:
        data_train.drop(col, axis=1, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler 
scale=MinMaxScaler()
data_train['scaled_column']=scale.fit_transform(data_train[['RecipeCode']])
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

X_train = data_train.drop(columns=['Rating'])
y_train = data_train['Rating']

X_test=pd.read_csv('/kaggle/input/recipe-for-rating-predict-food-ratings-using-ml/test.csv')
dummy_clf = DummyClassifier(strategy='most_frequent')  # 'most_frequent' strategy predicts the most frequent class
dummy_clf.fit(X_train, y_train)
y_pred = dummy_clf.predict(X_test)


```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_classifier_algo = KNeighborsClassifier(n_neighbors=5)

knn_classifier_algo.fit(X_train,y_train)
```

```python
#param_grid = {'n_neighbors': range(1, 15)} 
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors': range(1, 10)} 

grid_search = GridSearchCV(knn_classifier_algo, param_grid, cv=5, scoring='accuracy')
                           
grid_search.fit(X_train, y_train)
```

```python
k_grid = grid_search.best_params_['n_neighbors']
k_grid_score_best= grid_search.best_score_
```
