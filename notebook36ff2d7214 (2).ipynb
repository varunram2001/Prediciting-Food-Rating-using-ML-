```python
categorical_columns = data_tt.select_dtypes(include=['object','category']).columns.tolist()
print(categorical_columns)
```

```python
print("Unique recipies: ",data_tt['RecipeName'].nunique())
print("Unique CommentID", data_tt['CommentID'].nunique())
print('Unique userID', data_tt['UserID'].nunique())
print("Unique Usernames: ", data_tt['UserName'].nunique())
print("Unique Recipe_Review: ", data_tt['Recipe_Review'].nunique())

#13634 user comments which  is equal to number of rows in dataframe, so column can be dropped. 
```

```python
categorical_columns.remove('CommentID')
data_tt.drop(columns=['CommentID'], inplace=True)
data_val.drop(columns=['CommentID'], inplace=True)

```

```python
from sklearn.preprocessing import MinMaxScaler 
from sklearn.preprocessing import StandardScaler

scale1=MinMaxScaler()
scale2=StandardScaler()
```

```python
data_tt['RC_scaled_column']=scale1.fit_transform(data_tt[['RecipeCode']])
data_val['RC_scaled_column']=scale1.fit_transform(data_val[['RecipeCode']])

data_tt.drop(columns=['RecipeCode'], inplace=True)
data_val.drop(columns=['RecipeCode'], inplace=True)
```

```python
data_tt['UserReputation']=scale1.fit_transform(data_tt[['UserReputation']])
data_tt['ReplyCount']=scale1.fit_transform(data_tt[['ReplyCount']])
data_tt['ThumbsUpCount']=scale1.fit_transform(data_tt[['ThumbsUpCount']])
data_tt['ThumbsDownCount']=scale1.fit_transform(data_tt[['ThumbsDownCount']])
#data_tt['Rating']=scale1.fit_transform(data_tt[['Rating']])
data_tt['BestScore']=scale1.fit_transform(data_tt[['BestScore']])
```

```python
data_val['UserReputation']=scale1.fit_transform(data_val[['UserReputation']])
data_val['ReplyCount']=scale1.fit_transform(data_val[['ReplyCount']])
data_val['ThumbsUpCount']=scale1.fit_transform(data_val[['ThumbsUpCount']])
data_val['ThumbsDownCount']=scale1.fit_transform(data_val[['ThumbsDownCount']])
#data_val['Rating']=scale1.fit_transform(data_val[['Rating']])
data_val['BestScore']=scale1.fit_transform(data_val[['BestScore']])
```

```python
data_tt.head()
```

```python
data_val.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()

#Removing Recipe_Review alone as We will try to use tf vectorizer for that
categorical_columns.remove('Recipe_Review')

for i in categorical_columns:
    data_tt[i + '_L_Encoded'] = label_encoder.fit_transform(data_tt[i])
    data_val[i + '_L_Encoded'] = label_encoder.fit_transform(data_val[i])
    
    data_tt.drop(i, axis=1, inplace=True)
    
    data_val.drop(i, axis=1, inplace=True)

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()

tfidf.fit(data_tt['Recipe_Review'])


recipe_review_tfidf_train = tfidf.transform(data_tt['Recipe_Review'])
recipe_review_tfidf_val = tfidf.transform(data_val['Recipe_Review'])

tfidf_train_df = pd.DataFrame(recipe_review_tfidf_train.toarray(), columns=tfidf.get_feature_names_out())
tfidf_val_df = pd.DataFrame(recipe_review_tfidf_val.toarray(), columns=tfidf.get_feature_names_out())

# Concatenate TF-IDF features with original datasets
data_tt = pd.concat([data_tt.reset_index(drop=True), tfidf_train_df], axis=1)
data_val = pd.concat([data_val.reset_index(drop=True), tfidf_val_df], axis=1)
```

```python
data_tt.drop(columns=['Recipe_Review'], inplace=True)
data_val.drop(columns=['Recipe_Review'], inplace=True)
```
